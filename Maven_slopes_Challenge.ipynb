```python
df_merge=pd.merge(df_season_month,  df_snow[(df_season_month['resorts_Latitude'].between(df_snow['Latitude']-0.125, df_snow['Latitude']+0.125 )) & (df_season_month['resorts_Longitude'].between(df_snow['Longitude']-0.125, df_snow['Longitude']+0.125 ))] , how='left',left_on='SES_Num' , right_on='month_num')
```

```python
result=df_season_month.merge(df_snow,how='left',left_on='SES_Num' , right_on='month_num').loc[
    lambda df: (
        (df['resorts_Latitude'].between(df['Latitude'] - 0.125, df['Latitude'] + 0.125)) &
    (df['resorts_Longitude'].between(df['Longitude'] - 0.125, df['Longitude'] + 0.125))
    )].drop_duplicates()
result
```
